```python
scaler= preprocessing.StandardScaler()
x_normal=scaler.fit(data1).transform(data1)
x_normal
```

```python
k_mean=KMeans(init= 'k-means++', n_clusters=6, n_init=12)
k_mean.fit(x_normal)
```

```python
labels=k_mean.labels_
labels
```

```python
cluster_center=k_mean.cluster_centers_
cluster_center
```
